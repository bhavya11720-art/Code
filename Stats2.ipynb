```python
import pandas as pd
import numpy as np
```

1. Load the patient dataset and show summary with info()

```python
data=pd.read_csv('Patient_Data.csv')
data
```

```python
data.info()
```

2. Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
data_selected=data[['PatientID','Department','Doctor','BillAmount']]
print(data_selected)
```

3. Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
d=data.drop(['ReceptionistID','CheckInTime'],axis=1)
d
```

4. Use groupby to find total bill amount per department.

```python

data_1=d.groupby('Department').sum('BillAmount')
data_1
```

```python
data_2=data.groupby('Department').agg({'BillAmount': 'sum'})
data_2
```

5. Remove duplicate patient records based on PatientID.

```python
d=d.drop_duplicates('PatientID')
d
```

6. Fill missing BillAmount values with the mean bill amount.

```python
d=d.fillna(data.BillAmount.mean())
d
```

7. Merge the billing dataset with patient dataset on PatientID.


```python
data_b=pd.read_csv('Billing_Data.csv')
data_b
```

```python
d2=d.merge(data_b,on='PatientID')
d2
```

8.Concatenate an additional DataFrame that contains new patients for the current week (row-wise).

```python
new_data={'PatientID':[106,107],'Name':['Jasmin','Rahul'],'Department':['Cardiology','Dermatolgy'],'Doctor':['Dr.Jhon','Dr.Smith'],'BillAmount':[2000,1500]}
new_data
```

```python
new=pd.concat([d,pd.DataFrame(new_data)],ignore_index=True)
new
```

9. Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).

```python
dd={'InsuranceCovered':[1500,2500,3000,1500,2000,3500,2500],'FinalAmount':[4500,3500,4000,5500,6000,2500,3500]}
dd
```
